```python

import pandas as pd
df = pd.read_csv("../data_clean/aemet_obs_local_daily.csv", parse_dates=["fecha"])
df["fecha"] = pd.to_datetime(df["fecha"]).dt.floor("D")
df = df.sort_values(["fecha","idema"])
display(df.head(10))
print("Fechas:", df["fecha"].min().date(), "→", df["fecha"].max().date())
print("Estaciones:", df["idema"].nunique())
```

```python

import plotly.graph_objects as go, plotly.io as pio, os
pio.renderers.default = "vscode" if "VSCODE_PID" in os.environ else "notebook_connected"

res = df.groupby("fecha").agg(temp_media=("ta","mean"),
                              lluvia_total=("prec","sum"),
                              viento_medio=("vv","mean")).reset_index()

fig = go.Figure()
fig.add_trace(go.Scatter(x=res["fecha"], y=res["temp_media"], name="Temp media (°C)",
                         mode="lines", line=dict(color="red")))
fig.add_trace(go.Bar(x=res["fecha"], y=res["lluvia_total"], name="Lluvia (mm)",
                     marker_color="blue", opacity=0.5))
fig.add_trace(go.Scatter(x=res["fecha"], y=res["viento_medio"], name="Viento (m/s)",
                         mode="lines", line=dict(color="green"), yaxis="y2"))

fig.update_layout(title="AEMET (Dos Hermanas) — Resumen diario",
                  xaxis_title="Fecha",
                  yaxis_title="Temperatura / Lluvia",
                  yaxis2=dict(title="Viento", overlaying='y', side='right'),
                  legend=dict(orientation="h"))
fig.show()

# Guardar interactivo
fig.write_html("../reports/resumen_interactivo.html", include_plotlyjs="cdn")
print("✅ HTML -> ../reports/resumen_interactivo.html")
```
